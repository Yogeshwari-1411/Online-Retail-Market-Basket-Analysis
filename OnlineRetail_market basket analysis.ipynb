```python
top10=df['Country'].value_counts().head(10)
print(top10)
```

```python
plt.figure(figsize=[8,8])
plt.pie(top10,labels=top10.index,autopct='%0.0f%%',labeldistance=1.3)
plt.title("Distribution of Transaction by country")
plt.show()
```

```python
#converting data time accessor
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'],errors='coerce')
```

```python
time=df['InvoiceDate'].dt.year
df['year']=time
df.head()
```

```python
df.groupby('year')['UnitPrice'].sum().plot(kind='bar')
```

```python
df['Mon']=df["InvoiceDate"].dt.month
df['Month']=df["InvoiceDate"].dt.month_name()
df.groupby(['Mon','year'])['UnitPrice'].sum().plot(kind='pie',title='Month_wise Sales Report')
```

```python
df['Mon']=df["InvoiceDate"].dt.month
df['Month']=df["InvoiceDate"].dt.month_name()
df.groupby(['Mon','year'])['UnitPrice'].sum().plot(kind='bar',title='Month_wise Sales Report')
```
